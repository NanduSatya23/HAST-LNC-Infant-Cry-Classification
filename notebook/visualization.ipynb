# Model Performance Visualization

This notebook visualizes:
- Accuracy comparison
- ROC Curve
- Loss Curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
models = ["Logistic Regression", "MLP"]
accuracy = [0.89, 0.92]

plt.plot(models, accuracy, marker='o')
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.show()
```

```python
y_test_bin = np.random.randint(0, 2, 50)
y_scores = np.random.rand(50)

fpr, tpr, _ = roc_curve(y_test_bin, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

```python
loss_values = np.random.rand(50)

plt.plot(loss_values)
plt.title("Training Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
